<font color='CornflowerBlue'>**Matplotlib**</font> - основная библиотека визуализации двумерных данных на языке python.

Рисунок создается с помощью последовательного вызова функций из модуля **pyplot** (создание рисунка, задание области рисования, построение графиков в области рисования, украшение графика метками). Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

Вам дан двумерный непустой массив X размера n x m вещественных чисел.
Создайте на его основе массив состоящий: из каждого 4го (начиная с 0го и до конца) элемента по оси размерности n (axis=0), начиная с 120 до 500 (не включитально) с шагом 5 элементов по оси размерности m (axis=1).

Гарантируется, что m >= 500

```python
def get_part_of_array(X: List[List[float]]) -> List[List[float]]: # не векторизованная реализация
    """
    X - двумерный массив вещественных чисел размера n x m. Гарантируется что m >= 500
    Вернуть: двумерный массив, состоящий из каждого 4го элемента по оси размерности n
    и c 120 по 500 c шагом 5 по оси размерности m
    """
    res = []
    for line in X[::4]:
        res += [[i for i in line[120:500:5]]]
    return res


# X = np.array([list(np.arange(1, 501)) for i in range(5)])
# print(get_part_of_array(X))
```

```python
def get_part_of_array(X: np.ndarray) -> np.ndarray: # векторизованная
    """
    X - двумерный массив размера n x m. Гарантируется что m >= 500
    Вернуть: двумерный массив, состоящий из каждого 4го элемента по оси размерности n
    и c 120 по 500 c шагом 5 по оси размерности m
    """
    return X[::4, 120:500:5]

# X = np.array([list(np.arange(1, 501)) for i in range(5)])
# print(get_part_of_array(X))
```

#### <font color='DarkOrange'>**Задача 2 [0.3 + 0.7 баллов]**</font>

Вам дан двумерный непустой массив X целых чисел размера n x m.

Посчитайте сумму неотрицательных элементов на главной диагонали прямоугольной матрицы X. Если неотрицательных элементов на диагонали нет, то верните -1.

```python
def sum_non_neg_diag(X: List[List[int]]) -> int: # не векторизованная версия
    """
    Вернуть  сумму неотрицательных элементов на диагонали прямоугольной матрицы X.
    Если неотрицательных элементов на диагонали нет, то вернуть -1
    """
    res = []
    n = len(X)
    m = len(X[0])
    d = min(n, m)
    for i in range(d):
        if X[i][i] >= 0:
            res.append(X[i][i])
    if not res:
        return -1
    else:
        return sum(res)

# X = np.arange(-50, 50).reshape(10, 10)
# print(X)
# print(sum_non_neg_diag(X))
```

```python
def sum_non_neg_diag(X: np.ndarray) -> int: # векторизованная
    """
    Вернуть  сумму неотрицательных элементов на диагонали прямоугольной матрицы X.
    Если неотрицательных элементов на диагонали нет, то вернуть -1
    """
    d = X.diagonal()
    res = d[d >= 0]
    if res.size == 0:
        return -1
    else:
        return res.sum()

# X = np.arange(-50, 50).reshape(10, 10)
# print(X)
# print(sum_non_neg_diag(X))
```

Возможно пригодится функция deepcopy из модуля copy

```python
def replace_values(X: List[List[float]]) -> List[List[float]]:
    """
    X - двумерный массив вещественных чисел размера n x m.
    По каждому столбцу нужно почитать среднее значение M.
    В каждом столбце отдельно заменить: значения, которые < 0.25M или > 1.5M на -1
    Вернуть: двумерный массив, копию от X, с измененными значениями по правилу выше
    """
    n = len(X)
    m = len(X[0])
    new = deepcopy(X)
    meds = []
    for j in range(m):
        val = []
        for i in range(n):
            val += [new[i][j]]
        meds += [sum(val) / n]
    for j in range(m):
        m = meds[j]
        for i in range(n):
            if new[i][j] > 1.5*m or new[i][j] < 0.25*m:
                new[i][j] = -1
    return new

# X = np.ones((10, 10))
# X[0, :] = 100
# X[1, :] = 10
# print(X)
# print(replace_values(X))
```

```python
def replace_values(X: np.ndarray) -> np.ndarray:
    """
    X - двумерный массив вещественных чисел размера n x m.
    По каждому столбцу нужно почитать среднее значение M.
    В каждом столбце отдельно заменить: значения, которые < 0.25M или > 1.5M на -1
    Вернуть: двумерный массив, копию от X, с измененными значениями по правилу выше
    """
    y = np.copy(X)
    m = y.mean(axis=0) # среднее по столбцам
    return np.where(np.logical_or(X > 1.5*m, X < 0.25*m), -1, X)

# X = np.ones((10, 10))
# X[0, :] = 100
# X[1, :] = 10
# print(X)
# print(replace_values(X))
```
