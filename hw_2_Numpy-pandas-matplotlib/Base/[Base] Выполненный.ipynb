#### <font color='DarkOrange'>** 1 [0.3 + 0.7 ]**</font>

     X  n x m  .

```python
def get_part_of_array(X: List[List[float]]) -> List[List[float]]: #   
    """
    X -      n x m.   m >= 500
    :  ,    4     n
     c 120  500 c  5    m
    """
    res = []
    for line in X[::4]:
        res += [[i for i in line[120:500:5]]]
    return res


# X = np.array([list(np.arange(1, 501)) for i in range(5)])
# print(get_part_of_array(X))
```

```python
def get_part_of_array(X: np.ndarray) -> np.ndarray: # 
    """
    X -    n x m.   m >= 500
    :  ,    4     n
     c 120  500 c  5    m
    """
    return X[::4, 120:500:5]

# X = np.array([list(np.arange(1, 501)) for i in range(5)])
# print(get_part_of_array(X))
```

#### <font color='DarkOrange'>** 2 [0.3 + 0.7 ]**</font>

     X    n x m.

         X.      ,   -1.

```python
def sum_non_neg_diag(X: List[List[int]]) -> int: #   
    """
             X.
         ,   -1
    """
    res = []
    n = len(X)
    m = len(X[0])
    d = min(n, m)
    for i in range(d):
        if X[i][i] >= 0:
            res.append(X[i][i])
    if not res:
        return -1
    else:
        return sum(res)

# X = np.arange(-50, 50).reshape(10, 10)
# print(X)
# print(sum_non_neg_diag(X))
```

```python
def sum_non_neg_diag(X: np.ndarray) -> int: # 
    """
             X.
         ,   -1
    """
    d = X.diagonal()
    res = d[d >= 0]
    if res.size == 0:
        return -1
    else:
        return res.sum()

# X = np.arange(-50, 50).reshape(10, 10)
# print(X)
# print(sum_non_neg_diag(X))
```

#### <font color='DarkOrange'>** 1 [0.3 + 0.7 ]**</font>

    X    n x m.     ,      ,   1.5      1/4      -1

<font color='LightSteelBlue'>****</font>

   deepcopy   copy

```python
def replace_values(X: List[List[float]]) -> List[List[float]]:
    """
    X -      n x m.
           M.
        : ,  < 0.25M  > 1.5M  -1
    :  ,   X,      
    """
    n = len(X)
    m = len(X[0])
    new = deepcopy(X)
    meds = []
    for j in range(m):
        val = []
        for i in range(n):
            val += [new[i][j]]
        meds += [sum(val) / n]
    for j in range(m):
        m = meds[j]
        for i in range(n):
            if new[i][j] > 1.5*m or new[i][j] < 0.25*m:
                new[i][j] = -1
    return new

# X = np.ones((10, 10))
# X[0, :] = 100
# X[1, :] = 10
# print(X)
# print(replace_values(X))
```

```python
def replace_values(X: np.ndarray) -> np.ndarray:
    """
    X -      n x m.
           M.
        : ,  < 0.25M  > 1.5M  -1
    :  ,   X,      
    """
    y = np.copy(X)
    m = y.mean(axis=0) #   
    return np.where(np.logical_or(X > 1.5*m, X < 0.25*m), -1, X)

# X = np.ones((10, 10))
# X[0, :] = 100
# X[1, :] = 10
# print(X)
# print(replace_values(X))
```
