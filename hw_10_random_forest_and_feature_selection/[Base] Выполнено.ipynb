`````python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.datasets import fetch_california_housing
from sklearn.pipeline import Pipeline #  ,    Sklearn-knn-normalize-cv
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

#####
#      ,      
#       
# plt.style.use('tableau-colorblind10')
#####

import numpy as np

import time
import os

from itertools import combinations

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
`````

`````python
def get_california_dataset():
    """Return the splits of the California housing dataset for training and testing"""
    X, y = fetch_california_housing(return_X_y=True)
    return train_test_split(X, y, random_state=42)

X_train, X_test, y_train, y_test = get_california_dataset()
`````

`````python
from time import time

mae_rfs_test = []
mae_rfs_train = []
t = []
for n in np.logspace(1, 3, num=10):
    start = time()
    model = RandomForestRegressor(n_estimators=int(n), random_state=42, n_jobs=4)
    model.fit(X_train, y_train)
    mae_rfs_test.append(mean_absolute_error(model.predict(X_test), y_test))
    mae_rfs_train.append(mean_absolute_error(model.predict(X_train), y_train))
    t.append(time() - start)
`````

`````python
plt.figure(figsize=(12, 8))
plt.title(" MAE     ", fontsize=20)
plt.xlabel("  (n_estimators)")
plt.ylabel("MAE")
plt.xticks(np.arange(10), [int(i) for i in np.logspace(1, 3, num=10)])
plt.plot(np.arange(10), mae_rfs_train, label='train')
plt.plot(np.arange(10), mae_rfs_test, label='test')
plt.legend()
`````

`````python
def select_best_features(X_train, X_test, y_train, y_test, max_features=3):
    best_features = []
    best_ind = []
    feat = fetch_california_housing().feature_names
    for _ in range(max_features):
        mae = []
        for i in range(len(feat)):
            model = DecisionTreeRegressor()
            model.fit(X_train[:, np.array(best_ind + [i])], y_train)
            mae.append(mean_absolute_error(model.predict(X_test[:, np.array(best_ind + [i])]), y_test))
        best = np.argmin(mae)
        best_ind.append(best)
        best_features.append(feat[best])
        feat.remove(feat[best])
        
    return list(zip(best_ind, best_features))
`````

`````python
select_best_features(X_train, X_test, y_train, y_test)
`````
